#### Question 1 - What was the best month for sales and how much was earned that month

```python
# Removing na rows:
all_data_df.dropna(inplace=True)
```

```python
all_data_df
```

```python
# Removing irrelevant rows
filt = all_data_df['Order Date'] == 'Order'
irr_indexes = all_data_df.loc[filt].index
all_data_df.drop(irr_indexes, axis=0, inplace=True)
```

```python
all_data_df
```

```python
filt2 = all_data_df['Order Date'] == 'Order Date'
irr_index2 = all_data_df.loc[filt2].index
all_data_df.drop(irr_index2, axis=0, inplace=True)
```

```python
# Setting the date and time format
all_data_df['Order Date'] = pd.to_datetime(all_data_df['Order Date'], format = '%m/%d/%y %H:%M')
```

```python
# Create the month column from the Order Date column
all_data_df['Month'] = pd.DatetimeIndex(all_data_df['Order Date']).month
```

```python
# Changing the data type of the Quantity Ordered and Price Each column to float so that mathematical operations can be performed:
all_data_df['Quantity Ordered'] = all_data_df['Quantity Ordered'].astype('float')
all_data_df['Price Each'] = all_data_df['Price Each'].astype('float')
```

```python
# Creating a column of Total Sale:
all_data_df['Total Sale'] = all_data_df['Quantity Ordered']*all_data_df['Price Each']
```

```python
# Grouping the dataframe by Months:
grouped_df = all_data_df.groupby('Month')
```

```python
all_data_df['Total Sale'] = all_data_df['Total Sale'].astype('float')
```

#### QUESTION 2:- What city has the highest no of sales?

```python
# Splitting the address into separate columns:
all_data_df['Purchase Address'].str.split(",")
```

```python
# Creating new columns for St Name, City, State:
all_data_df[['St Name', 'City', 'State']] = all_data_df['Purchase Address'].str.split(",", expand=True)
```

```python
def sep_state(state_add):
    state = state_add[0:3] 
    return state
all_data_df['City'] = all_data_df['City'] + all_data_df['State'].apply(lambda x: sep_state(x))
```

```python
all_data_df
```

```python
# Grouping the df by cities and summing the values of all columns to get the sum of Total Sale (and also the sum of Month column which is irrelevant date)
city_grp = all_data_df.groupby('City').sum()
```

```python
city_grp = city_grp.reset_index()
```

```python
sale = city_grp['Total Sale']
```

```python
cities = city_grp['City']
```

```python
plt.bar(cities, sale)
plt.ylabel('Sales', color = 'white')
plt.xticks(cities, rotation='vertical')
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
```

```python
# ANSWER 2 :- The city of San Fransico has the highest amount of sales for the the year
```

#### QUESTION 4:- What products are often sold together?

```python
all_data_df['Grouped'] = all_data_df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
grouped_df = all_data_df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
grouped_df
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for rows in grouped_df['Grouped']:
    row_list = rows.split(',')
    count.update(Counter(combinations(row_list,2)))

for key, value in count.most_common(10):
    print(key, value)
    
```

QUESTION 6 - WHICH PRODUCT GENERATED THE MOST REVENUE?

```python
quantity_ordered_df = pd.DataFrame(quantity_ordered)
```

```python
quantity_ordered_df.reset_index(inplace=True)
```

```python
quantity_ordered_df['Price Each'] = all_data_df['Price Each']
quantity_ordered_df['Price Each'].astype(float)
```

```python
# Calculating the total sales figures for each product:-
quantity_ordered_df['Total SP'] = quantity_ordered_df['Quantity Ordered'] * quantity_ordered_df['Price Each']
```

```python
most_revenue = quantity_ordered_df.sort_values(by='Total SP', ascending=False)
most_revenue
```

```python
plt.bar(most_revenue['Product'], most_revenue['Total SP'])
plt.xticks(rotation='vertical', color='white')
plt.yticks(color='white')
plt.show()
```
